# Load in data

```python
exp4572 = pd.read_csv("Mass_spec_exp4572.csv", sep=";")
exp4557 = pd.read_csv("/groups/gerlich/experiments/Experiments_004500"
                      "/004557/Mass_spec_data.csv", sep=";")
exp4557.columns = ["Name", "Value", "Rep"]
```

```python
fin = pd.concat((exp4572, exp4557), axis=0)
```

```python
mean = fin.groupby("Name").mean().reset_index()
```

```python
f, ax = plt.subplots()
sbn.barplot(x="Name", y="Value", data=mean, ax=ax, color="black",facecolor=(1, 1, 1, 0) , linewidth=2,edgecolor="black", order=["Control", "4sT"])
sbn.stripplot(x="Name", y="Value", data=fin, color="black", size=10, jitter=0.1, ax=ax, order=["Control", "4sT"])
ax.set_xlabel("")
ax.set_ylabel("Percent 4sT [%]")
sbn.despine()
plt.xticks(*plt.xticks(), rotation=90)
sbn.despine()
ax.set_ylim(0, 3.1)
f.set_size_inches(4, 6)
plt.show()
```

```python
f.set_size_inches(4, 6)
f.savefig("./4sT_exp4557_exp4572_black_wPoints_indSamples.png", bbox_inches="tight")
f.savefig("./4sT_exp4557_exp4572_black_wPoints_indSamples.pdf", bbox_inches="tight")
```
